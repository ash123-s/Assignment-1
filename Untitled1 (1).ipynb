```
weight=float (input("enter the weight in pounds"))
height=float (input("enter thr height in inches"))
wkilo=weight*0.45359237
hemeter=height*0.0254
bmi=wkilo/(hemeter**2)
if (bmi<18.5):
  print("UNDERWEIGHT")
elif (bmi<=18.5 or bmi<25.0):
  print("NORMAL")
elif (bmi<=25.0 or bmi<30.0):
  print("OVERWEIGHT")
else:
  print("OBESE")
```
